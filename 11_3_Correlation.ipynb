## Correlation magnitude and sign

- Correlation is a number between -1 and 1.  A value of 0 means that two numbers are NOT correlated
- A value of -1 demonstrates a perfect negative linear association
- A value of 1 is a perfect positive linear association

## Correlation vs. Causation (differences, use cases)

- Here we're looking at values that are correlated.  This makes no assumption of a causal relationship
- Causal relationships are best determined through hypothesis testing, creating a controlled setting to determine a causal relationship

## Correlation in Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
df = pd.read_csv('employee_attrition.csv')
```
